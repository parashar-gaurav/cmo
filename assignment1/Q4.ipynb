```python
Q = np.matrix([[2.3346,1.1384,2.5606,1.4507],[1.1384,0.7860,1.2743,0.9531],[2.5606,1.2743,2.8147,1.6487],[1.4507,0.9531,1.6487,1.8123]])
```

```python
b = np.matrix([0.4218,0.9157,0.7922,0.9595]).T
```

```python
x_0 = np.matrix([-135.1150,-4.5224,130.1168,-5.6879]).T
#x_0 = np.matrix([-135.4797,-4.5251,130.4600,-5.7051]).T
```

##### Trying if x* = $Q^{-1}b$ 
##### Calculating $\frac{1}{2}x^TQx-b^Tx$ and $\frac{1}{2}x^{*T}Qx^{*}-b^Tx^{*}$

```python
x_opt = np.matmul(np.linalg.inv(Q),b)
print(x_opt)
```

```python
f_opt = 0.5*(np.matmul(np.matmul(x_opt.T,Q),x_opt)) - np.matmul(b.T,x_opt)
f_x = 0.5*(np.matmul(np.matmul(x_0.T,Q),x_0)) - np.matmul(b.T,x_0)
print(f_opt)
```

```python
def l2_norm(a,b):
    return np.sqrt(np.sum(np.power(a-b,2)))

def l1_norm(a,b):
    return np.sum(np.abs(a-b))

def grad_f(x):
    res = np.matmul(Q,x) - b
    return res
```

```python
alpha_1 = 0.0728
alpha_2 = 0.2185
```

##### part (c)

```python
z = x_0
iters = 0
while(l2_norm(z,x_opt)>0.0001):
    z = z - alpha_1*grad_f(z)
    iters += 1
print("For alhpa 1, number of iterations = ",iters)

z = x_0
iters = 0
while(l2_norm(z,x_opt)>0.0001):
    z = z - alpha_2*grad_f(z)
    iters += 1
print("For alhpa 2, number of iterations = ",iters)
```
