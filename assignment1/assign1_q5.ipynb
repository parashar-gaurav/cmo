# GRADIENT DESCENT

```python
def l2_norm(x):
    return np.sqrt(np.sum(np.power(x,2)))
```

# NEWTON'S METHOD

##### $x_{k+1} = x_k - H^{-1}(x_k)\nabla f(x_k)$

```python
# part m
z = x_0
eps = 0.1
iters = 0
grad_z = np.matrix(f.rosen_2d_grad(z)).T
print("initial l2 norm of grad_z = ",l2_norm(grad_z))
while l2_norm(grad_z) > eps:
    hess_z_inv = np.linalg.inv(np.matrix(f.rosen_2d_Hess(z)).astype(float))
    second_term = np.matmul(hess_z_inv,grad_z)
    z = z - second_term
    grad_z = np.matrix(f.rosen_2d_grad(z)).T
    iters+=1
    #print("iteration = ",iters,"norm of grad_z = ",l2_norm(grad_z))
print("total number of iterations = ", iters)
```
