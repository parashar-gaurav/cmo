```python
import pickle
import dill
import numpy as np
```

```python
with open('f6.pkl','rb') as f:
    data = pickle.load(f)
f6 = dill.loads(data)

with open('grad_f6.pkl','rb') as f:
    data = pickle.load(f)
grad_f6 = dill.loads(data)
```

```python
x_0 = [10,100,100,10]
```

```python

def l2_norm(x1):
    x1 = np.matrix(x1).T
    res = np.sqrt(np.sum(np.power(x1,2)))
    return res

def dotProd(lista,listb):
    res = [a*b for a,b in zip(lista,listb)]
    return sum(res)

def inexact_ls(x_init,f,g,alpha,beta,eps):
    x_k = x_init
    iters = 0
    t = 1
    call_f6 = 0
    call_grad_f6 = 0
    grad = g(x_k)
    call_grad_f6 += 1
    u = [-1*element for element in grad]
    tu = [t*elem for elem in u]
    while (l2_norm(tu)>=eps):
        iters = iters+1
        print('iteration number = ',iters)
        t = 1

        alpha_t = alpha*t
        call_f6 += 1
        while((f([sum(n) for n in zip(*[x_k, tu])]))>(f(x_k)+alpha*t*dotProd(grad,u))):
            #print('1st function value = ',f([sum(n) for n in zip(*[x_k, tu])]))
            #print('2nd function value = ',f(x_k)+alpha_t*dotProd(grad,u))
            call_f6 += 1
            t = beta*t
            alpha_t = alpha*t
            tu = [t*elem for elem in u]
        print('best t found for iteration = ',iters)
        call_f6 += 1
        
        x_k = [sum(n) for n in zip(*[x_k, tu])]

        print('l2_norm of tu * 1e+7',l2_norm(tu)*1e+7)

    print('least number of calls to f6 = ',call_f6)
    print('least number of calls to grad_f6 = ',call_grad_f6)
    return x_k,iters   
```

```python
x_0 = [10,100,100,10]
x_opt,iterations = inexact_ls(x_0,f6,grad_f6,0.5,0.5,1e-7)
print(x_opt)
print('iterations = ',iterations)
```
