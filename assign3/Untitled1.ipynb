```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.power(x-5,4)
def grad_f(x):
    return 4*np.power(x-5,3)
def hess_f(x):
    return 12*np.power(x-5,2)
```

```python
x_list = []
y_list = []
grad_list = []
hess_list = []
start = -100
end = 6
x = 4
while x<=end:
    x_list.append(x)
    y_list.append(f(x))
    grad_list.append(grad_f(x))
    hess_list.append(hess_f(x))
    x+=0.01
```

```python
plt.plot(x_list,y_list)
#plt.plot(x_list,grad_list)
#plt.plot(x_list,hess_list)
```

```python
def NewMe(x,k):
    iters = 0
    x_old = x
    x_axis = []
    y_axis = []
    while iters <k:
        iters+=1
        x_axis.append(iters)
        x_new = x_old - (grad_f(x_old)/hess_f(x_old))
        res = (abs(x_new - 5))/(abs(x_old-5))
        print('x_value = ',x_new)
        print('iterations = ',iters)
        print('numerator = ',abs(x_new-5))
        print('denum = ',abs(x_old-5))
        print('res = ',res)
        x_old = x_new
        y_axis.append(res)
    plt.plot(x_axis,y_axis)
    
```

```python
NewMe(10,10000)
```
