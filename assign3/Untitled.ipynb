```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    if x<0:
        return np.power(-np.power(abs(x),1/3),4)
    return np.power(np.power(abs(x),1/3),4)

def grad_f(x):
    if x<0:
        return (4/3)*(-np.power(abs(x),1/3))
    return (4/3)*(np.power(abs(x),1/3))

def hess_f(x):
    if x<0:
        return (4/9)*(np.power(abs(1/x),2/3))
    return (4/9)*(np.power((1/x),2/3))
```

```python
x_list = []
y_list = []
y_grad_list = []
y_hess_list = []
start = -10
end = 10
x = start
while x<end:
    if x!=0:
        x_list.append(x)
        y_list.append(f(x))
        y_grad_list.append(grad_f(x))
        y_hess_list.append(hess_f(x))
    x+=0.5
```

```python
plt.plot(x_list,y_list)
plt.plot(x_list,y_grad_list)
plt.plot(x_list,y_hess_list)
```

```python
def newton_method(x_init):
    x_old = x_init
    iters = 0
    while(abs(x_old - 0)>=1e-6):
        iters+=1
        print('iters = ',iters)
        x_new = x_old - (grad_f(x_old)/hess_f(x_old))
        print('x_old = ', x_old, 'x_new = ',x_new)
        x_old = x_new
```

```python
newton_method(1e-5)
```
