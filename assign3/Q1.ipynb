```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 100*np.power((x[1]-x[0]*x[0]),2) + np.power((1 - x[0]),2)
def grad_f(x):
    res = []
    res.append(-400*x[0]*x[1]+400*x[0]*x[0]*x[0]-2+2*x[0])
    res.append(200*x[1]-200*x[0]*x[0])
    return np.array(res)

def hess_f(x):
    res = np.matrix([[0,0],[0,0]])
    res[0,0] = -400*x[1]+1200*x[0]*x[0]+2
    res[0,1] = -400*x[0]
    res[1,0] = -400*x[0]
    res[1,1] = 200
    return res
```

```python
hess_f([100,100])
```

# Applying Newton method

```python
def NewMe(x,k):
    iters = 0
    x_old = x
    while iters <k:
        iters+=1
        print('iterations = ',iters)
        print('x_k = ', x_old)
        print('gradient x_k', grad_f(x_old))
        print('multiplication ',np.matmul(np.linalg.inv(hess_f(x_old)),np.matrix(grad_f(x_old)).T))
        print('inverse :',np.linalg.inv(hess_f(x_old)))
        x_new = np.matrix(x_old).T - np.matmul(np.linalg.inv(hess_f(x_old)),np.matrix(grad_f(x_old)).T)
        #print('iteration = ',iters,'x = ',x_new)
        x_old[0] = x_new[0,0]
        x_old[1] = x_new[1,0]
        #print('iteration = ',iters,'x = ',x_old)
        print('hessian = ',hess_f(x_old))
        print('gradient = ',grad_f(x_old))
        
```

```python
NewMe([0,0],2)
```

```python
NewMe([100,100],160)
```
